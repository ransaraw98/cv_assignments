```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img_orig = cv.imread('../images/forest.jpg', cv.IMREAD_COLOR)
print(img_orig.shape)

cv.namedWindow("Image", cv.WINDOW_AUTOSIZE)
cv.imshow("Image", img_orig)
cv.waitKey(0)
gamma =2
table = np.array([(i/255)**(gamma)*255 for i in np.arange(0,256)])
img_gamma  = cv.LUT(img_orig, table)

cv.namedWindow("Gamma Corrected", cv.WINDOW_AUTOSIZE)
cv.imshow("Gamma Corrected", img_gamma)

cv.waitKey(0)
```
