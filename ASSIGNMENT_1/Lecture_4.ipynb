```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

img = cv.imread('../images/gal_gadot.jpg', cv.IMREAD_REDUCED_GRAYSCALE_2)

sigma =2
kernel = cv.getGaussianKernel(5, sigma)
blurred = cv.sepFilter2D(img, -1, kernel, kernel, anchor =(-1,1), delta = 0, borderType = cv.BORDER_REPLICATE)
diff = img.astype('float32') - blurred.astype('float32')

sharpened = cv.addWeighted(img.astype('float32'), 1.0, diff, 1.5, 0)

fig, axes = plt.subplots(2,2,sharex='all', sharey = 'all', figsize=(15,15))
axes[0,0].imshow(sharpened, cmap = 'gray')
plt.show()
```

```python
import cv2 as cv
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

img = cv.imread('../images/gal_gadot.jpg', cv.IMREAD_REDUCED_GRAYSCALE_2)

kernel = np.array([(0,-1,0),(-1,5,-1),(0,-1,0)], dtype ='float')
imgc = cv.filter2D(img,-1,kernel)

fig, axes = plt.subplots(1,2,sharex='all', sharey = 'all', figsize=(18,10))
axes[0].imshow(img, cmap ='gray')
axes[0].set_title('Original')
#axes[0].set_xticks([]), axes[0].set_ytickes([])
axes[1].imshow(imgc, cmap ='gray')
axes[1].set_title('Sharpened')
#axes[1].set_xticks([]), axes[1].set_ytickes([])
plt.show()
```
