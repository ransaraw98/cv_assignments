```python
%matplotlib inline
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
gal = cv2.imread("../images/gal_gadot.jpg", cv.IMREAD_GRAYSCALE)
```

```python
np.shape(gal)
```

```python
hist_R = cv2.calcHist([gal],[0],None, [256],[0,256])
hist_G = cv2.calcHist([gal],[1],None, [256],[0,256])
hist_B = cv2.calcHist([gal],[2],None, [256],[0,256])
```

```python
fig, axes = plt.subplots(1,3,squeeze = False,figsize = (20,7), sharex = True)
axes[0,0].plot(hist_R,'r')
axes[0,0].set_ylabel('# of pixels')
axes[0,1].plot(hist_G,'g')
axes[0,2].plot(hist_B, 'b')
axes[0,0].set_xlabel("Intensity")
axes[0,1].set_xlabel("Intensity")
axes[0,2].set_xlabel("Intensity")
plt.plot()
plt.show()
```
