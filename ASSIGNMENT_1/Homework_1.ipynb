```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
img = cv.imread('grey.png',0)
ret,thresh1 = cv.threshold(img,127,255,cv.THRESH_BINARY)
ret,thresh2 = cv.threshold(img,127,255,cv.THRESH_BINARY_INV)
ret,thresh3 = cv.threshold(img,127,255,cv.THRESH_TRUNC)
ret,thresh4 = cv.threshold(img,127,255,cv.THRESH_TOZERO)
ret,thresh5 = cv.threshold(img,127,255,cv.THRESH_TOZERO_INV)
titles = ['Original Image','BINARY','BINARY_INV','TRUNC','TOZERO','TOZERO_INV']
images = [img, thresh1, thresh2, thresh3, thresh4, thresh5]
for i in range(6):
    plt.subplot(2,3,i+1),plt.imshow(images[i],'gray',vmin=0,vmax=255)
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```

```python
%matplotlib inline
import cv2 as cv 
import matplotlib.pyplot as plt
imgo = cv.imread('../images/gal_gadot.jpg',1)
imgc = cv.cvtColor(imgo, cv.COLOR_BGR2RGB)
imgray = cv.imread('../images/gal_gadot.jpg',0)
fig, ax = plt.subplots(1,2)
ax[0].imshow(imgray,cmap='gray')
ax[0].set_title('Without color conversion')
ax[0].axis('off')
ax[1].imshow(imgc)
ax[1].set_title('With color conversion')
ax[1].axis('off')
plt.show()
```

```python
%matplotlib inline
import cv2 as cv
import matplotlib.pyplot as plt
img = cv.imread('../images/random_chick.jpg',cv.IMREAD_COLOR)
img = cv.cvtColor(img,cv.COLOR_BGR2RGB)
img_b = img +100
img_c = cv.add(img,100)
fig,ax = plt.subplots(1,3)
fig.set_size_inches(18.5, 10.5)
ax[0].imshow(img)
ax[0].set_title('untouched')
ax[0].axis('off')
ax[1].imshow(img_b)
ax[1].set_title('img +100')
ax[1].axis('off')
ax[2].imshow(img_c)
ax[2].set_title('cv.add')
ax[2].axis('off')
plt.show()
```
